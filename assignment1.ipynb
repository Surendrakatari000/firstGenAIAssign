```python
!pip install transformers

# Step 2: Import the pipeline function
from transformers import pipeline

# Step 3: Load a non-gated model using pipeline for text generation
# Using 'distilgpt2' as it is a smaller, non-gated model for demonstration
generator = pipeline("text-generation", model="distilgpt2")

# Step 4: Provide your prompt
prompt = "Explain quantum computing in simple terms."

# Step 5: Generate text using the model
output = generator(prompt, max_new_tokens=100, do_sample=True, temperature=0.7)

# Step 6: Print the result
print(output[0]['generated_text'])
```

```python
# Install the Transformers library
!pip install transformers
from transformers import pipeline

generator = pipeline("text2text-generation", model="google/flan-t5-base")

#  Zero-shot Prompting
zero_shot = "Translate English to Hindi: I love programming."
print("\n🔹 Zero-shot Result:")
print(generator(zero_shot)[0]["generated_text"])


#  Few-shot Prompting
few_shot = """Translate English to Hindi:

English: Hello
Hindi: नमस्ते

English: How are you?
Hindi: आप कैसे हैं?

English: I love programming.
Hindi:"""

print(" Few-shot Result:")
print(generator(few_shot)[0]["generated_text"])


#  Chain-of-Thought Prompting
cot_prompt = """Q: Tom has 3 apples. He buys 2 more. Then he gives 1 to his friend. How many apples does Tom have now?
A: Let's think step by step."""

print("Chain-of-Thought Result:")
print(generator(cot_prompt)[0]["generated_text"])


#  Prompt Template (Dynamic)
template = "Summarize the following text in one sentence:\n\n{text}"
article = "OpenAI released a new model called GPT-4o. It can handle text, image, and audio simultaneously. It is faster and cheaper than GPT-4."
filled_prompt = template.format(text=article)

print("Prompt Template Result:")
print(generator(filled_prompt)[0]["generated_text"])


# Role-based Instructional Prompt (System-style - works okay in FLAN)
role_prompt = "You are a helpful assistant. Explain what AI is in simple words."
print("Role-style Prompt Result:")
print(generator(role_prompt)[0]["generated_text"])
```
